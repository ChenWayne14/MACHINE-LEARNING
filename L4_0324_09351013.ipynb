```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data['age'].plot.hist(bins=50)
```

```python
DC = pd.crosstab(data['survived'],data['sex'])
DC
```

```python
DC.plot()
```

```python
data.groupby('survived')['age'].plot(kind='hist', bins=30, alpha=0.5, legend=True)
```

```python
sns.boxplot(x="sex",y='age',hue="survived",data=data,palette="coolwarm")
```

```python
sns.heatmap(data.corr(),cmap='coolwarm',annot=True)
```

```python
sns.lmplot(x='age',y='fare',data=data)
```

```python
sns.jointplot(x='age',y='fare',data=data,kind='scatter')
```
